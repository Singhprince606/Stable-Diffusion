```python
# =============================
# Text-to-Image Converter (Interactive Prompt)
# Academic Project - Google Colab
# =============================

# Step 1: Install required libraries
!pip install "diffusers[torch]" transformers accelerate --quiet
!pip install git+https://github.com/huggingface/diffusers --quiet

# Step 2: Import dependencies
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import datetime

# Step 3: Load the pre-trained model
pipe = StableDiffusionPipeline.from_pretrained(
    "redstonehero/cetusmix_v4",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")
pipe.safety_checker = None  # Disable safety checker for safe prompts

# Step 4: Ask user for input prompt
prompt = input("Enter your text prompt: ")

# Negative prompt to reduce quality issues (preset)
negative_prompt = (
    "easynegative, human, lowres, bad anatomy, bad hands, text, error, missing fingers, "
    "extra digit, cropped, worst quality, low quality, watermark, blurry, bad feet, "
    "poorly drawn hands, mutation, deformed, extra limbs, malformed limbs, long neck"
)

# Generation parameters
height = 800
width = 640
steps = 25
guidance = 7.5

# Step 5: Generate the image
image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    height=height,
    width=width,
    num_inference_steps=steps,
    guidance_scale=guidance
).images[0]

# Step 6: Show and save the image
image.show()
display(image)

# # Optional: Save image to file with timestamp
# timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
# image.save(f"generated_image_{timestamp}.png")
# print(f"Image saved as generated_image_{timestamp}.png")
```
